```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift , estimate_bandwidth

centers = [(2,2),(2,6),(7,4)]
n_classes = len(centers)

x , targets = make_blobs(n_samples = 10000 , centers= centers, cluster_std =0.5)

bandwidth = estimate_bandwidth(x, quantile=0.2, n_samples=500)

ms = MeanShift(bandwidth = bandwidth, bin_seeding = True)
ms.fit(x)

labels = ms.labels_
unique_labels = np.unique(labels)
n_clusters = len(unique_labels)
cluster_center =ms.cluster_centers_
print('Number of unique clusters = %d' %n_clusters)

print('Cluster Centers :')
print(cluster_center)
pred = ms.predict(x)
#print(x)

colors = list(map(lambda x: 'r' if x==1 else 'g' if x==2 else 'y',pred))
plt.scatter(x[:,0],x[:,1], c=colors, marker ='o', picker =True)
plt.show()



```
